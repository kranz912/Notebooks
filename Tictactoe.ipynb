```python
import pandas as pd
import json
from pandas.io.json import json_normalize 
import hvplot.pandas
```

```python
with open('Datafiles/tictactoe1000.json','r') as output:
    data =json.load(output)
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.columns
```

```python
json_normalize(df['1'])
```

```python
for x in df.columns:
    df[x] = json_normalize(df[x])[['fitness']]
```

```python
df['1']
```

```python
df
```

```python
df=  df.transpose()
df= df.replace(-3,'O')
df= df.replace(3,'X')
df= df.replace(0,'Draw')
```

```python
df
```

```python
df
```

```python
df = df.transpose()
```

```python
tally = df.apply(pandas.Series.value_counts)
```

```python
tally = tally.transpose()
```

```python
tally
```

```python
tally.index.name ='generation'
```

```python
tally= tally.reset_index()
```

```python
tally['generation'] = pd.to_numeric(tally['generation'])
```

```python
tally['generation']
```

```python
tally.hvplot(x='generation', y=['X','O','Draw'],kind='scatter',figsize='200,200')
```
